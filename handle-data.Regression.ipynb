**Simple Linear Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
df = pd.DataFrame({"Hours":[2,4,6,8,10], "Marks":[50,60,65,70,75]})
X = df[["Hours"]]
y = df["Marks"]

# Fit SLR model
model = LinearRegression()
model.fit(X, y)

# Predict
y_pred = model.predict(X)


print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
r2 = model.score(X, y)
print("Coefficient of Determination (R²):", r2)
      
plt.scatter(X, y, color="blue", label="Actual data")  # actual points
plt.plot(X, y_pred, color="red", label="Regression line")  # regression line
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data
df = pd.DataFrame({
    "Area":[1000,1500,2000],
    "Bedrooms":[2,3,4],
    "Age":[10,5,2],
    "Price":[300000, 450000, 600000]
})

# Features and target
X = df[["Area", "Bedrooms", "Age"]]
y = df["Price"]

# Fit model
model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Predict
y_pred = model.predict(X)

# Plot predicted vs actual
plt.scatter(y, y_pred, color="blue")
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', label="Perfect prediction")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Multiple Linear Regression: Actual vs Predicted")
plt.legend()
plt.show()
```
