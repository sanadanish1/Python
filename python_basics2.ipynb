```python
score = input("Enter Score: ")
s = float(score)

if 0.0>= s >= 1.0:
  print("score out of range")
else:
    if s>=0.9:
        print("A")
    elif s>=0.8:
        print("B")
    elif s>=0.7:
        print("C")
    elif s>=0.6:
        print("D")
    elif s<0.6:
        print("F")
```

```python
largest = None
smallest = None
while True:
    num = input("Enter a number: ")
    try:
        num == int(num)
        if largest is None or num < largest:
            largest = num
        if smallest is None or num > largest:
            smallest = num
    except:
        print("Invalid input")
    if num == "done":
        break
if largest is not None:
    print("Maximum is", largest)
if smallest is not None:
    print("Minimum is", smallest)
```
