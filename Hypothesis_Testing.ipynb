## One-Sample t-Test (for mean)

Use when:

Population standard deviation is unknown

```python
# 1. Import libraries
from scipy import stats
import pandas as pd
import numpy as np

# 2. Create Sample Data
data = pd.read_excel("Book1.xlsx")
sample=data['sugar']
sample_mean=np.mean(sample)

# 3. Define Hypotheses
# H0: Hypothesized (population) mean = 0
# H1: population mean != 0 (two-tailed test)
mu_0 = 0

# 4. Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(sample, popmean=mu_0)

print("t-statistic:", t_stat)
print("p-value:", p_value)

# 5. Decision rule
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")


# 6. Confidence Interval
confidence_level = 1 - alpha
n=len(sample)
ci = stats.t.interval(confidence_level, df=n-1, loc=sample_mean, scale=stats.sem(sample))

print("\nConfidence Interval:", ci)

# 7. Interpretation
print("\nInterpretation:")
print("If p-value < 0.05, we conclude that the population mean is significantly different from", mu_0)
print("Otherwise, we do not have enough evidence to reject the claim.")
```

```python
# Daily sales data
sales = [102, 98, 190, 97, 105, 103, 56]
sales_mean=np.mean(sales)
# Hypothesized mean
mu_0 = 0
t_stat, p_value=stats.ttest_1samp(sales, popmean=mu_0)

print(sales_mean)
print("t-statistic:", t_stat)
print("p-value:", p_value)

#decision rule
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")

# 7. Confidence Interval
confidence_level = 1 - alpha
n=len(sales)
ci = stats.t.interval(confidence_level, df=n-1, loc=sales_mean, scale=stats.sem(sales))

print("\nConfidence Interval:", ci)
```

## Right Tail Test

```python
import numpy as np
from scipy import stats

# Daily sales data
sales = [102, 98, 190, 97, 105, 103, 56]

sales_mean = np.mean(sales)

# Hypothesized mean
mu_0 = 0

t_stat, p_value_two = stats.ttest_1samp(sales, popmean=mu_0)

# Convert to one-tailed p-value
p_value_one = p_value_two / 2

print("Mean:", sales_mean)
print("t-statistic:", t_stat)
print("One-tailed p-value:", p_value_one)

# Decision rule
alpha = 0.05

if (t_stat > 0) and (p_value_one < alpha):
    print("Reject H0 (Right-tailed test)")
else:
    print("Fail to Reject H0")
    
# Left tail case
# p_value_one = p_value_two / 2

# if (t_stat < 0) and (p_value_one < alpha):
#     print("Reject H0 (Left-tailed test)")
# else:
#     print("Fail to Reject H0")

# One-sided Confidence Interval (Lower bound)
n = len(sales)
confidence_level = 1 - alpha

ci_lower = stats.t.ppf(alpha, df=n-1, loc=sales_mean, scale=stats.sem(sales))

print("\nOne-sided CI (Lower bound):", ci_lower)
```

## Two Sample t-Test (for mean)

### Independent samples (unequal variances)

Seperate variance t-test

```python
import numpy as np
from scipy import stats

# Two groups (example: sales of two stores)
group1 = [102, 98, 190, 97, 105, 103, 56]
group2 = [95, 85, 100, 90, 92, 88, 91]

mean1 = np.mean(group1)
mean2 = np.mean(group2)

# Two-sample t-test (Welch by default)
t_stat, p_value_two = stats.ttest_ind(group1, group2, equal_var=False, alternative='greater')

# Convert to one-tailed p-value
p_value_one = p_value_two / 2

print("Mean Group1:", mean1)
print("Mean Group2:", mean2)
print("t-statistic:", t_stat)
print("One-tailed p-value:", p_value_one)

# Decision rule
alpha = 0.05

if (t_stat > 0) and (p_value_one < alpha):
    print("Reject H0 (Group1 > Group2)")
else:
    print("Fail to Reject H0")
```

pooled variance t-test

```python
import numpy as np
from scipy import stats

# Two groups (example: sales of two stores)
group1 = [102, 98, 190, 97, 105, 103, 56]
group2 = [95, 85, 100, 90, 92, 88, 91]

mean1 = np.mean(group1)
mean2 = np.mean(group2)

# Two-sample t-test (Welch by default)
t_stat, p_value_two = stats.ttest_ind(group1, group2, equal_var=True, alternative='greater')

# Convert to one-tailed p-value
p_value_one = p_value_two / 2

print("Mean Group1:", mean1)
print("Mean Group2:", mean2)
print("t-statistic:", t_stat)
print("One-tailed p-value:", p_value_one)

# Decision rule
alpha = 0.05

if (t_stat > 0) and (p_value_one < alpha):
    print("Reject H0 (Group1 > Group2)")
else:
    print("Fail to Reject H0")
```

### Related samples 

Paired t-test

```python
import numpy as np
from scipy import stats

# Two groups (example: sales of two stores)
group1 = [102, 98, 190, 97, 105, 103, 56]
group2 = [95, 85, 100, 90, 92, 88, 91]

mean1 = np.mean(group1)
mean2 = np.mean(group2)

# Two-sample t-test (Welch by default)
t_stat, p_value_two = stats.ttest_rel(group1, group2)

# Convert to one-tailed p-value
p_value_one = p_value_two / 2

print("Mean Group1:", mean1)
print("Mean Group2:", mean2)
print("t-statistic:", t_stat)
print("One-tailed p-value:", p_value_one)

# Decision rule
alpha = 0.05

if (t_stat > 0) and (p_value_one < alpha):
    print("Reject H0 (Group1 > Group2)")
else:
    print("Fail to Reject H0")
```
