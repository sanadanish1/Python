**Dictionary**

Collection of objects of different types stored in the key-value pair format enclosed within {}, each value sperated from its key by colon :

**Syntax**

Dict1=[1:'a', 2:'b', 3:'c'}

**Characteristics:**

- It has mutable nature
- Dup;icates are not allowed
- Elements can't be acessed by index


```python
#create empty dict
Dict1={}
Dict1={'income':"false", 'consumption':'true'}
print(Dict1['income'])
```

```python
# Python code to demonstrate the working of median() on various range of data-sets

from statistics import median
from fractions import Fraction as fr

# tuple of positive integer numbers
data1 = (10, 20, 30, 40, 50)

# tuple of floating point values
data2 = (2.4, 5.1, 6.7, 8.9)

# tuple of fractional numbers
data3 = (fr(1, 2), fr(44, 12),
        fr(10, 3), fr(2, 3))

# tuple of a set of negative integers
data4 = (-5, -1, -12, -19, -3)

# tuple of set of positive and negative integers
data5 = (-1, -2, -3, -4, 4, 3, 2, 1)

# Printing the median of above datasets
print("Median of data-set 1 is % s" % (median(data1)))
print("Median of data-set 2 is % s" % (median(data2)))
print("Median of data-set 3 is % s" % (median(data3)))
print("Median of data-set 4 is % s" % (median(data4)))
print("Median of data-set 5 is % s" % (median(data5)))
```

```python
import statistics

data = [2, 4, 5, 2, 3, 2, 4, 5, 2]

mode = statistics.mode(data)
print("Mode:", mode)
```

```python
from statistics import multimode

data = [1, 2, 2, 3, 3, 4]

modes = multimode(data)
print("Modes:", modes)
```

```python
from collections import Counter

data = [2, 4, 5, 2, 3, 2, 4, 5, 4]

counts = Counter(data)
mode = counts.most_common(1)[0][0]

print("Mode:", mode)
```
