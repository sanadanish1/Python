```python
# Python code to demonstrate the working of median() on various range of data-sets

from statistics import median
from fractions import Fraction as fr

# tuple of positive integer numbers
data1 = (10, 20, 30, 40, 50)

# tuple of floating point values
data2 = (2.4, 5.1, 6.7, 8.9)

# tuple of fractional numbers
data3 = (fr(1, 2), fr(44, 12),
        fr(10, 3), fr(2, 3))

# tuple of a set of negative integers
data4 = (-5, -1, -12, -19, -3)

# tuple of set of positive and negative integers
data5 = (-1, -2, -3, -4, 4, 3, 2, 1)

# Printing the median of above datasets
print("Median of data-set 1 is % s" % (median(data1)))
print("Median of data-set 2 is % s" % (median(data2)))
print("Median of data-set 3 is % s" % (median(data3)))
print("Median of data-set 4 is % s" % (median(data4)))
print("Median of data-set 5 is % s" % (median(data5)))
```

```python
import numpy as np

# Example dataset
data = np.array([4, 5, 6, 4, 4, 7, 6, 5, 4])

# Find unique values and their counts
values, counts = np.unique(data, return_counts=True)

# Find the index of the maximum count
max_count_index = np.argmax(counts)

# Mode value
mode_value = values[max_count_index]

print("Mode of the data:", mode_value)
```

```python
import numpy as np

# Example dataset (multimodal: 4 and 5 both appear most frequently)
data = np.array([4, 5, 6, 4, 5, 7, 6, 5, 4])

# Get unique values and their counts
values, counts = np.unique(data, return_counts=True)

# Find the highest frequency
max_count = np.max(counts)

# Get all values with that frequency (all modes)
modes = values[counts == max_count]

print("Modes of the data:", modes)
```

```python
from statistics import mode
mode_value = mode([8, 9, 9, 10, 9, 8, 10])
print("Mode:", mode_value)
```

Standard Deviation:
Square root of variance ‚Äî shows how spread out data is.

üí° Example:
In weather data, a small std dev means temperatures stay close to average,
a large std dev means temperatures vary a lot.

```python
std_value = np.std([45, 50, 55])
print("Standard Deviation:", std_value)
```
