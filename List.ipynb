**List**

List is a data structure in which we can store positionally ordered collection of arbitrary object types like integer, float, string etc.

- It has mutable nature
- It has no fixed size
- Duplicates are allowed


**Syntax**

list_name=[Item 1, Item 2, ..., Item n]

```python
#create empty list
list1=[]
print(type(list1))
print(list1)


#list using list function
list2=list()
print(type(list2))
```

```python
#create a list and print it
list_1=['alpha', 'beta', 'gamma']
print(list_1)
```

**Operations on list**

**Find length of the list**

```python
#find length of list
print(len(list_1))
```

**Add element to the list**

```python

#method 1: "append" automatically adds at last position (you can append another list to this list)
list_1.append(["a", "b"])

#method 2: add element at specific index
#list_1.insert(2, "mu")

print(list_1)
```

**Replace the element**

```python

list_1[1]=15
print(list_1)
```

**To count how many times a specific item is present**

```python

print(list_1.count("mu"))
```

**To undo append or delete element**

pop:pop modifies the list and returns the element that was removed

#If you don’t provide an index, it deletes and returns the last element

```python


#method 1: to del with item name
#list_1.remove("mu")

#method 2: del by index
#del list_1[-1]

#method 3: delete last element or at specific index
#list_1.pop()
list_1.pop(1)

print(list_1)
```

**To find index of specific element**

```python
#list_1.index('beta')
#call item by index
print(list_1[1])
```

**To arrange items**

```python
#arrange in ascending order
#list_1.sort()

#to arrange in descending or reverse order
list_1.sort(reverse=True)
print(list_1)
```

**slicing specific region inside list**

```python
#from index 3 to 5
#list_1[3:6]

#from index 2 to last element
list_1[2:]

#from start to index 3
#list_1[:4]
```

**cancatenating two or more lists**
