**Simple Linear Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
df = pd.read_excel("Book1.xlsx")
X = df[["sugar"]]
Y = df["calories"]

# Fit SLR model
model = LinearRegression()
model.fit(X, Y)

# Predict
Y_pred = model.predict(X)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
print("Coefficient of Determination (R²):", model.score(X, Y))
      
plt.scatter(X, Y, color="blue", label="Actual data")  # actual points
plt.plot(X, Y_pred, color="red", label="Regression line")  # regression line
plt.xlabel("sugar")
plt.ylabel("calories")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
df = pd.DataFrame({"Hours":[2,4,6,8,10], "Marks":[50,60,65,70,75]})
X = df[["Hours"]]
y = df["Marks"]

# Fit SLR model
model = LinearRegression()
model.fit(X, y)

# Predict
y_pred = model.predict(X)


print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
r2 = model.score(X, y)
print("Coefficient of Determination (R²):", r2)
      
plt.scatter(X, y, color="blue", label="Actual data")  # actual points
plt.plot(X, y_pred, color="red", label="Regression line")  # regression line
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```
